```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import re
import pandas as pd 
import requests
import os
```

```python
def get_chrome_driver():

    chrome_options = webdriver.ChromeOptions()

    driver = webdriver.Chrome(
        service = Service(ChromeDriverManager().install()),
        options = chrome_options

    )
    return driver
```

```python
# driver 생성
driver = get_chrome_driver()
NH_URL = "https://smartmarket.nonghyup.com/servlet/content/nhbank/pd/sfpd0101c.thtml"
driver.get(NH_URL)
time.sleep(3)
try:
    link = driver.find_element(By.PARTIAL_LINK_TEXT, '펀드상품/약관')
    link.click()
except Exception as e:
    print("오류 발생: ", e)
time.sleep(3)
page = driver.page_source
# html 구조 확인
soup = BeautifulSoup(page, 'html.parser')

```

```python
# icons = soup.find_all(class_='btn_table type1')
def pdf_URL(icons):
    urls = [icon['onclick'] for icon in icons]
    # 정규 표현식을 사용하여 URL만 추출
    url_pattern = re.compile(r"common_getFileDownload\('(.*?)'")
    urls_extracted = [url_pattern.search(url).group(1) for url in urls]
    
    # URL을 리스트 형식으로 저장
    url_list = []
    for url in urls_extracted:
        url_list.append(url)
    
    return url_list

# 함수 호출하여 URL을 리스트로 저장
pdf_urls = pdf_URL(icons)
print(pdf_urls)
```

```python
# "상품설명서"로 끝나는 문자열 찾기 위한 정규 표현식 패턴
pattern = re.compile(r'\b상품설명서\b')

# 클래스 't_left'를 가진 모든 요소 찾기
Name = soup.find_all(class_='t_left')

#패턴에 맞는 요소의 index 번호만 출력
def indexing(Name, pattern):
    number = []
    for index, element in enumerate(Name):
        if pattern.search(element.get_text()):
            number.append(index)
    icons_matched = [pdf_urls[i] for i in number]
    return icons_matched

indexing(Name, pattern)
```

```python
#국내 펀드 URL 뽑기 
driver = get_chrome_driver()
NH_URL = "https://smartmarket.nonghyup.com/servlet/content/nhbank/pd/sfpd0101c.thtml"
driver.get(NH_URL)
time.sleep(3)
link = driver.find_element(By.PARTIAL_LINK_TEXT, '펀드상품/약관')
link.click()
time.sleep(3)
collected_urls = []  # 추출된 URL을 저장할 리스트
product_ids = ["01","02", "04", "05", "07", "08", "09", "10", "11", "12", "98"]
for product_id in product_ids:
    try:
        # 상품 페이지로 이동
        element = driver.find_element(By.ID, f"li_{product_id}")
        driver.execute_script("arguments[0].click();", element)
        time.sleep(3)
        
        
        page_number = 4  # 초기 페이지 버튼 인덱스
        
        # '다음' 페이지가 없을 때까지 반복
        while True:
            try:
                page = driver.page_source
                soup = BeautifulSoup(page, 'html.parser')
                
                # 페이지에서 URL 추출하는 로직
                icons = soup.find_all(class_='btn_table type1')
                pdf_urls = pdf_URL(icons)
                Name = soup.find_all(class_='t_left')
                icons_matched = indexing(Name, pattern)
                collected_urls.extend(icons_matched)
                
                # '다음' 페이지 버튼 찾기 및 클릭
                next_button_xpath = '//*[@id="paging"]/span[13]/a'
                next_button = driver.find_element(By.XPATH, next_button_xpath)
                if next_button:
                    next_button.click()
                    time.sleep(3)
                else:
                    break  # '다음' 버튼이 없으면 반복 종료
            
            except Exception as e:
                print(f"마지막 페이지 또는 오류 발생: {e}")
                break
        
        # 추출된 URL 출력 또는 저장
        print(f"Product ID {product_id} collected URLs:", collected_urls)
    except Exception as e:
        print(f"상품 ID {product_id}에서 오류 발생: {e}")
```

```python
# 지정한 저장 디렉토리 경로
save_dir = 'C:\\Users\\USER\\Desktop\\세미나\\project\\농협은행_펀드상품_pdf'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
# df의 URL 컬럼을 순회
for i, row in df.iterrows():
    url = row['URL']  # 'url_column_name'을 실제 URL 컬럼 이름으로 교체
    try:
        response = requests.get(url, timeout=10)  # 10초 타임아웃 설정
        response.raise_for_status()  # 상태 코드가 200이 아니면 예외를 발생시킴

        # 파일명 추출 (예: '0325700MN001_pi.pdf')
        filename = url.split('/')[-1]

        # 파일 저장 경로 설정
        file_path = os.path.join(save_dir, filename)

        # PDF 파일 저장
        with open(file_path, 'wb') as f:
            f.write(response.content)

        print(f'{filename} 다운로드 완료')
    except requests.exceptions.RequestException as e:
        print(f'{url} 다운로드 실패: {e}')

print('모든 파일 다운로드 완료')
```
