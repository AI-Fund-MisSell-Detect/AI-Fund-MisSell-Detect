```python
import os
import shutil
import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
options = ChromeOptions()
#options.add_argument("headless")

#기존 쿠키를 이용하여 드라이버 사용
options.add_argument("--user-data-dir=/Users/subin/User Data")
driver = webdriver.Chrome(options=options)

# 기존 다운로드 경로와 새 다운로드 경로 정의
old_download_path = '/Users/subin/Downloads'
new_download_path = '/Users/subin/Desktop/subin/유신투자증권'

# 설명서와 약관의 인덱스 변수 선언
manual_index = 1
terms_index = 1

for i in range(1, 6):
    # 크롬 드라이버에 url 주소 넣고 실행
    driver.get(f'https://www.eugenefn.com/ingo/igfd/igfd100r.do?menu_id=02030100&menu_url=%2Fingo%2Figfd%2Figfd100r.do&menu_level=3&page_id=02030100&page_url=%2Fingo%2Figfd%2Figfd100r.do&page_level=3&m_fundCd=&m_fundNm=&m_f_risk=&m_clsNm=&m_f_pa_gubun=&m_f_elf_cd=&m_dftp_able_yn=&m_ottp_able_yn=&m_rsp_able_yn=&s_detailGubun=%25&s_detailGubunNm=%EC%A0%84%EC%B2%B4&s_companyCode=%25&s_companyName=%EC%A0%84%EC%B2%B4&searchGubun=1&areaGubun=&foreignGubun=&detailGubun=&riskGubun=&trwGubun=&areaGubunVal=&foreignGubunVal=&detailGubunVal=&riskGubunVal=&trwGubunVal=&totalcnt=669&pageno={i}&pageSize=20&fnd_pd_ivst_rsk_clsf_cod=&fundCode=&fundCd=&sort_turn_cd=2&sort_ctn_cd=1Y&inputValue=&returnUrl=&fnd_type=&onlineYn=Y&showAll=Y&105p_invsType=&e2e_data1=&fnd_fit_nm=&fundName=&d_saveGubun=&ernGubun=%25&ernSta=&ernEnd=&setSta=&setEnd=&online=on&pageNo=34')
    # 페이지가 완전히 로딩되도록 기다림
    time.sleep(random.randint(2, 5))

    for j in range(1, 21):
        fundname = driver.find_element(By.XPATH, f'//*[@id="contents"]/div[2]/div[5]/div[1]/table/tbody/tr[{2*j-1}]/td[2]/a')
        fundname.click()
        time.sleep(random.randint(2, 5))

        # 설명서 다운로드
        manual = driver.find_element(By.XPATH, '//*[@id="fund_contents"]/div[3]/div[2]/div[2]/ul/li[3]/a')
        manual.click()
        time.sleep(5)  # 다운로드 완료를 확인하기 위한 대기 시간

        # 파일 이름 변경
        filename = max([old_download_path + '/' + f for f in os.listdir(old_download_path)], key=os.path.getctime)
        new_filename = f"설명서 {manual_index}.pdf"
        os.rename(os.path.join(old_download_path, filename), os.path.join(old_download_path, new_filename))

        # 파일 이동
        shutil.move(os.path.join(old_download_path, new_filename), new_download_path)

        # 설명서 인덱스 증가
        manual_index += 1

        # 약관 다운로드
        terms = driver.find_element(By.XPATH, '//*[@id="fund_contents"]/div[3]/div[2]/div[2]/ul/li[4]/a')
        terms.click()
        time.sleep(5)  # 다운로드 완료를 확인하기 위한 대기 시간

        # 파일 이름 변경
        filename = max([old_download_path + '/' + f for f in os.listdir(old_download_path)], key=os.path.getctime)
        new_filename = f"약관 {terms_index}.pdf"
        os.rename(os.path.join(old_download_path, filename), os.path.join(old_download_path, new_filename))

        # 파일 이동
        shutil.move(os.path.join(old_download_path, new_filename), new_download_path)

        # 약관 인덱스 증가
        terms_index += 1

        driver.back() # 이전 페이지로 이동
        time.sleep(random.randint(2, 5))
        driver.refresh() # 페이지 새로 고침
```
