```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
```

```python
# driver 생성
driver = get_chrome_driver()
KB_URL = "https://obank.kbstar.com/quics?page=C020003#loading"
driver.get(KB_URL)
```

```python
def download(driver, start_page, end_page):
    current_page = start_page
    while current_page <= end_page:
        # 현재 페이지의 다운로드 링크를 클릭
        for i in range(1, 11):  # 각 페이지에서 1부터 10까지의 다운로드 링크
            try:
                xpath = f'//*[@id="b034261"]/form/table/tbody/tr[{i}]/td[2]/span/span/button'
                download_link = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, xpath))
                )
                download_link.click()

                # 경고창 처리
                try:
                    alert = driver.switch_to.alert
                    alert.accept()
                except Exception:
                    # 경고창이 없는 경우, 이곳에서 예외가 발생하고 무시
                    pass
            except Exception as e:
                print(f"An error occurred: {e}")

        # 다음 페이지
        if current_page % 10 == 0:  # 10, 20, 30... 페이지에서는 다음 페이지 그룹으로 이동
            next_page_group_link = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="nextinput"]/span/button'))
            )
            next_page_group_link.click()
        else:  # 그 외 페이지에서는 다음 페이지로 이동
            next_page_link = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, f'//*[@id="pageinput{current_page + 1}"]/span/input'))
            )
            next_page_link.click()

        current_page += 1
```

```python
## 상품설명서 또는 규약 페이지로 이동 후 실행해야함
# 1페이지 부터 2페이지까지
download(driver, 1, 2)
```
